```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt

import sklearn
df4 = pd.read_csv ('jo11.csv')
print (df4.columns)
```

```python
def get_timediff(m1,m2,d1,d2,h1,h2,q1,q2,s1,s2):
    seconds = 2629746*(m1-m2) + 86400*(d1-d2) + 3600*(h1-h2) + 60*(q1-q2) + (s1-s2)
    return seconds

for i in range(0,966):
   df4.at[i+1,'seconds']=get_timediff(df4.at[i,'month'],df4.at[i+1,'month'],df4.at[i,'date'],df4.at[i+1,'date'],df4.at[i,'hour'],df4.at[i+1,'hour'],df4.at[i,'minutes'],df4.at[i+1,'minutes'],df4.at[i,'seconds'],df4.at[i+1,'seconds'])
#print()
   if df4.at[i+1,'seconds']<0:
         df4.at[i+1,'seconds']=- df4.at[i+1,'seconds']
        
```
